```python
import pandas as pd
```

```python
races = pd.read_csv('races.csv')
```

```python
races = races[races['year'] >= 2014]
```

```python
races = races.reset_index()
```

```python
races
```

```python
results = pd.read_csv('results.csv')
```

```python
results
```

```python
race_results = races.merge(results, left_on='raceId', right_on='raceId')
```

```python
race_results
```

```python
drivers = pd.read_csv('drivers.csv')
```

```python
drivers
```

```python
data =  race_results.merge(drivers, left_on='driverId', right_on='driverId')
```

```python
lap = pd.read_csv('lap_times.csv')
```

```python
data = data.merge(lap, left_on='raceId', right_on='raceId')
```

```python
data = data.drop(columns=['time_x','url_x','fastestLapTime','fastestLapSpeed','url_y'], axis= 1)
```

```python
data = data.drop(columns=['number_x','number_y','dob','nationality','code','fastestLap'], axis= 1)
```

```python
display(data.tail(3))
display(data.head(3))
```

```python
data1 = data.groupby(['driverId_x','raceId','circuitId','name','driverRef','grid','constructorId','positionOrder']).agg({'laps':'max'})
```

```python
data1 = data1.reset_index()
```

```python
data1.head(5)
```

```python
con = pd.read_csv('constructors.csv')

```

```python
data3 = data1.merge(con,left_on='constructorId',right_on='constructorId')

data3.head(2)
```

```python
data2 = data3.drop(['driverId_x','raceId','laps','url','name_x','constructorId','nationality','name_y'], axis=1)
```

```python
data2.head(2)
```

```python
circuit = pd.read_csv('circuits.csv')

final = data2.merge(circuit,left_on='circuitId',right_on='circuitId')

final.head(2)
```

```python
##create clean df

final.to_csv('f1(1).csv')
```
